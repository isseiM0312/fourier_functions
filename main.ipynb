```python
import numpy as np
import matplotlib.pyplot as plt

# 
def f(x):
    return np.where((x > -np.pi) & (x < 0), 1, -1)

# 
def a0():
    return 0

def an(n):
    return 0

def bn(n):
    if n % 2 == 0:
        return 0
    else:
        return 4 / (np.pi * n)

# 
def fourier_series(x, n_terms):
    a0_term = a0() / 2
    series = a0_term * np.ones_like(x)
    for n in range(1, n_terms + 1):
        series += an(n) * np.cos(n * x) + bn(n) * np.sin(n * x)
    return series

# 
x = np.linspace(-2 * np.pi, 2 * np.pi, 1000)
y = f(x)
n_terms = 10  # 

# 
plt.figure(figsize=(12, 6))
plt.plot(x, y, label='Original Function $f(x)$')
plt.plot(x, fourier_series(x, n_terms), label=f'Fourier Series with {n_terms} terms')

# 
discontinuity_points = [-np.pi, 0, np.pi]
for point in discontinuity_points:
    plt.axvline(point, color='gray', linestyle='--', alpha=0.5)

plt.title('Convergence of Fourier Series for a Piecewise Smooth Function')
plt.xlabel('$x$')
plt.ylabel('$f(x)$ and Fourier Series')
plt.legend()
plt.grid(True)
plt.show()
```
